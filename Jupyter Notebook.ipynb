En este ejemplo, la clase Circulo tiene un método constructor que toma el radio como parámetro y lo asigna como un atributo de la instancia (self.radio). También tiene propiedades (obtener_area y obtener_circunferencia) que calculan el área y la circunferencia del círculo, respectivamente.

El método imprimir_informacion imprime información sobre el círculo, incluyendo su radio, área y circunferencia.

La demostración de uso crea una instancia de la clase Circulo llamada circulo1 y muestra cómo acceder a sus atributos, propiedades y llamar a su método.

```python
import math

class Circulo:
    # Método constructor
    def __init__(self, radio):
        self.radio = radio

    # Propiedad para obtener el área
    @property
    def obtener_area(self):
        return math.pi * self.radio**2

    # Propiedad para obtener la circunferencia
    @property
    def obtener_circunferencia(self):
        return 2 * math.pi * self.radio

    # Método para imprimir información sobre el círculo
    def imprimir_informacion(self):
        print(f"Círculo con radio: {self.radio}")
        print(f"Área: {self.obtener_area}")
        print(f"Circunferencia: {self.obtener_circunferencia}")

# Demostración de uso
# Crear una instancia de la clase Circulo
circulo1 = Circulo(radio=5)

# Acceder a atributos
print(f"Radio: {circulo1.radio}")

# Acceder a propiedades
print(f"Área: {circulo1.obtener_area}")
print(f"Circunferencia: {circulo1.obtener_circunferencia}")

# Llamar al método imprimir_informacion
circulo1.imprimir_informacion()
```
